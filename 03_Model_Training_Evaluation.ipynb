# AI Fake News Detection System

## Model Training and Evaluatio

```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

## Load Cleaned Dataset

```python
# Load cleaned dataset

df = pd.read_csv("cleaned_data.csv")

print("Cleaned dataset loaded successfully.")
```

```python
# Display dataset shape

print(df.shape)
```

```python
# Display first five records

df.head()
```

```python
# Display dataset information

df.info()
```

```python
# Check missing values

df.isnull().sum()
```

## Define Features and Target

```python
# Define input feature and target variable

X = df["clean_text"]
y = df["label"]

print("Features and target variable created successfully.")
```

```python
# Display feature and target shapes

print("Feature Shape :", X.shape)
print("Target Shape  :", y.shape)
```

## TF-IDF Vectorization

```python
# Import TF-IDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Create TF-IDF Vectorizer

tfidf = TfidfVectorizer(
    max_features=10000,
    ngram_range=(1,2),
    min_df=2
)

print("TF-IDF Vectorizer created successfully.")
```

## Apply TF-IDF Vectorization

```python
# Remove missing values from clean_text
df["clean_text"] = df["clean_text"].fillna("")

# Select input text
X = df["clean_text"]
```

```python
X = tfidf.fit_transform(X)
print("TF-IDF vectorization completed successfully.")
```

```python
# Transform text data into TF-IDF features
#X = tfidf.fit_transform(X)
#print("TF-IDF vectorization completed successfully.")
```

```python
# Display TF-IDF matrix shape

print("TF-IDF Shape:", X.shape)
```

## Train-Test Split

```python
# Import train_test_split

from sklearn.model_selection import train_test_split
```

```python
# Split dataset into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Train-Test split completed successfully.")
```

```python
# Display training and testing shapes

print("X_train :", X_train.shape)
print("X_test  :", X_test.shape)
print("y_train :", y_train.shape)
print("y_test  :", y_test.shape)
```

## Logistic Regression Model

```python
# Import Logistic Regression

from sklearn.linear_model import LogisticRegression
```

```python
# Train Logistic Regression model

lr_model = LogisticRegression(max_iter=1000)

lr_model.fit(X_train, y_train)

print("Logistic Regression model trained successfully.")
```

```python
# Predict on test data

lr_predictions = lr_model.predict(X_test)
```

```python
# Import evaluation metrics

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
# Evaluate Logistic Regression model

print("Accuracy  :", accuracy_score(y_test, lr_predictions))
print("Precision :", precision_score(y_test, lr_predictions))
print("Recall    :", recall_score(y_test, lr_predictions))
print("F1 Score  :", f1_score(y_test, lr_predictions))
```

## Multinomial Naive Bayes Model

```python
# Import Multinomial Naive Bayes

from sklearn.naive_bayes import MultinomialNB
```

```python
# Train Multinomial Naive Bayes model
nb_model = MultinomialNB()

nb_model.fit(X_train, y_train)

print("Multinomial Naive Bayes model trained successfully.")
```

```python
# Predict on test data

nb_predictions = nb_model.predict(X_test)
```

```python
# Evaluate Multinomial Naive Bayes model

print("Accuracy  :", accuracy_score(y_test, nb_predictions))
print("Precision :", precision_score(y_test, nb_predictions))
print("Recall    :", recall_score(y_test, nb_predictions))
print("F1 Score  :", f1_score(y_test, nb_predictions))
```

## Decision Tree Model

```python
# Import Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
```

```python
# Train Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)

print("Decision Tree model trained successfully.")
```

```python
# Predict on test data

dt_predictions = dt_model.predict(X_test)
```

```python
# Evaluate Decision Tree model

print("Accuracy  :", accuracy_score(y_test, dt_predictions))
print("Precision :", precision_score(y_test, dt_predictions))
print("Recall    :", recall_score(y_test, dt_predictions))
print("F1 Score  :", f1_score(y_test, dt_predictions))
```

## Random Forest Model

```python
# Import Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
```

```python
# Train Random Forest model

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

```python
# Predict on test data

rf_predictions = rf_model.predict(X_test)
```

```python
# Evaluate Random Forest model

print("Accuracy  :", accuracy_score(y_test, rf_predictions))
print("Precision :", precision_score(y_test, rf_predictions))
print("Recall    :", recall_score(y_test, rf_predictions))
print("F1 Score  :", f1_score(y_test, rf_predictions))
```

## Model Comparison

```python
# Create comparison table

comparison = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Multinomial Naive Bayes",
        "Decision Tree",
        "Random Forest"
    ],
    "Accuracy": [
        accuracy_score(y_test, lr_predictions),
        accuracy_score(y_test, nb_predictions),
        accuracy_score(y_test, dt_predictions),
        accuracy_score(y_test, rf_predictions)
    ],
    "Precision": [
        precision_score(y_test, lr_predictions),
        precision_score(y_test, nb_predictions),
        precision_score(y_test, dt_predictions),
        precision_score(y_test, rf_predictions)
    ],
    "Recall": [
        recall_score(y_test, lr_predictions),
        recall_score(y_test, nb_predictions),
        recall_score(y_test, dt_predictions),
        recall_score(y_test, rf_predictions)
    ],
    "F1 Score": [
        f1_score(y_test, lr_predictions),
        f1_score(y_test, nb_predictions),
        f1_score(y_test, dt_predictions),
        f1_score(y_test, rf_predictions)
    ]
})

comparison.sort_values(by="Accuracy", ascending=False)
```

```python
# Display the best model

best_model = comparison.sort_values(
    by="Accuracy",
    ascending=False
).iloc[0]

print(best_model)
```

## Support Vector Machine (Linear SVC)

```python
# Import LinearSVC

from sklearn.svm import LinearSVC
```

```python
# Train LinearSVC model

svm_model = LinearSVC(random_state=42)

svm_model.fit(X_train, y_train)

print("LinearSVC model trained successfully.")
```

```python
# Predict on test data

svm_predictions = svm_model.predict(X_test)
```

```python
# Evaluate LinearSVC model

print("Accuracy  :", accuracy_score(y_test, svm_predictions))
print("Precision :", precision_score(y_test, svm_predictions))
print("Recall    :", recall_score(y_test, svm_predictions))
print("F1 Score  :", f1_score(y_test, svm_predictions))
```

```python
# Add LinearSVC results to comparison table

comparison.loc[len(comparison)] = [
    "LinearSVC",
    accuracy_score(y_test, svm_predictions),
    precision_score(y_test, svm_predictions),
    recall_score(y_test, svm_predictions),
    f1_score(y_test, svm_predictions)
]

comparison = comparison.sort_values(by="Accuracy", ascending=False)

comparison
```

```python
# Display the best model

print("Best Performing Model")
comparison.head(1)
```

## Confusion Matrix

```python
# Import confusion matrix

from sklearn.metrics import confusion_matrix
```

```python
# Create confusion matrix for the best model (LinearSVC)

cm = confusion_matrix(y_test, svm_predictions)

cm
```

```python
# Visualize confusion matrix

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Fake", "Real"],
    yticklabels=["Fake", "Real"]
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")

plt.tight_layout()
plt.show()
```

## Classification Report

```python
# Import classification report

from sklearn.metrics import classification_report
```

```python
# Display classification report

print(classification_report(y_test, svm_predictions))
```

## Accuracy Comparison

```python
# Visualize model accuracy comparison
plt.figure(figsize=(9,5))

ax = sns.barplot(
    data=comparison,
    x="Model",
    y="Accuracy"
)

plt.title("Model Accuracy Comparison", fontsize=14, fontweight="bold")
plt.xlabel("Machine Learning Model")
plt.ylabel("Accuracy")

for container in ax.containers:
    ax.bar_label(container, fmt="%.4f")

plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
# Test the loaded model

sample_news = [
    "The government announced a new policy to improve education."
]

sample_vector = loaded_tfidf.transform(sample_news)

prediction = loaded_model.predict(sample_vector)

if prediction[0] == 0:
    print("Prediction : Fake News")
else:
    print("Prediction : Real News")
```
